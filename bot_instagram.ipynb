```python
# Bibliotecas
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from PyQt5 import QtCore, QtGui, QtWidgets, uic
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QMainWindow, QProgressBar, QLineEdit
from PyQt5.QtWidgets import QMessageBox, QLabel

import time
import random

# webdriver
spyder = webdriver.Edge()
spyder.get("https://www.instagram.com/")
time.sleep(5)
# Funções

    # reset do login
def Reset():
    tela.line_email.clear()
    tela.line_senha.clear()
    
    # Login
def Logar():
    email = 'infosistemas.py'#tela.line_email.text()
    senha = '21417692'#tela.line_senha.text()
    time.sleep(2)
    
    spyder.find_element(By.XPATH, '/html/body/div[1]/section/main/article/div[2]/div[1]/div[2]/form/div/div[1]/div/label/input').send_keys(email)
    time.sleep(1)
    spyder.find_element(By.XPATH, '/html/body/div[1]/section/main/article/div[2]/div[1]/div[2]/form/div/div[2]/div/label/input').send_keys(senha)
    time.sleep(2)
    spyder.find_element(By.XPATH, '/html/body/div[1]/section/main/article/div[2]/div[1]/div[2]/form/div/div[3]/button/div').click()
    time.sleep(3)
    spyder.find_element(By.XPATH, '/html/body/div[1]/section/main/div/div/div/div/button').click()
    time.sleep(4)
    spyder.find_element(By.XPATH, '/html/body/div[6]/div/div/div/div[3]/button[2]').click()
    time.sleep(4)
    Troca_controle()
    time.sleep(2)
    
def Troca_controle():
    Reset()
    controle.show()
    tela.close()
    Pega_nome()
    time.sleep(3)
    
def Troca_tela():
    spyder.get("https://www.instagram.com/")
    tela.show()
    controle.close()
    
def Pega_nome():
    n_conta = spyder.find_element(By.XPATH, '/html/body/div[1]/section/main/section/div[3]/div[1]/div/div/div[2]/div[2]/div/div/div').text()
    controle.label_nome_conta.setText(str(n_conta))
    
    # Seguir 
    
    # Curtir

def comente_nas_fotos_com_a_hashtag(spyder, hashtag):
        links_de_posts = []
        spyder = webdriver.Edge()
        spyder.get("https://www.instagram.com/explore/tags/" + hashtag + "/")
        time.sleep(5)
        for i in range(
            1, 3
        ):  # Altere o segundo valor aqui para que ele desça a quantidade de páginas que você quiser: quer que ele desça 5 páginas então você deve alterar de range(1,3) para range(1,5)
            spyder.execute_script(
                "window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(3)
        hrefs = spyder.find_element(By.NAME, "a")
        pic_hrefs = [elem.get_attribute("href") for elem in hrefs]
        print(hashtag + " fotos: " + str(len(pic_hrefs)))
        for link in pic_hrefs:
            try:
                if link.index("/p/") != -1:
                    links_de_posts.append(link)
            except:
                pass

        for pic_href in links_de_posts:
            spyder.get(pic_href)
            spyder.execute_script(
                "window.scrollTo(0, document.body.scrollHeight);")
            try:
                comments = [
                    "Olha que foto massa!",
                    "Olha que foto massa!",
                    "Olha que foto massa!",
                    "Olha que foto massa!",
                ]  # Remova esses comentários e insira os seus comentários aqui
                spyder.find_element(By.CLASS_NAME, "Ypffh").click()
                comment_input_box = spyder.find_element(By.CLASS_NAME, "Ypffh")
                time.sleep(random.randint(2, 5))
                self.type_like_a_person(
                    random.choice(comments), comment_input_box)
                time.sleep(random.randint(3, 5))
                spyder.find_element(By.XPATH, "//button[contains(text(), 'Publicar')]").click()
                time.sleep(random.randint(3, 5))
            except Exception as e:
                print(e)
                time.sleep(5)

app=QtWidgets.QApplication([])
tela=uic.loadUi('tela_login.ui')
controle=uic.loadUi('controle.ui')

# Metodos

    # tela_login
tela.botao_limpar.clicked.connect(Reset)
tela.botao_logar.clicked.connect(Logar)

   # painel de controle


controle.botao_sair.clicked.connect(Troca_tela)
controle.botao_curtir.clicked.connect(comente_nas_fotos_com_a_hashtag)

#tela.lineEdit.text()

#tela.pushButton_4.clicked.connect()
#tela.pushButton_5.clicked.connect()
#nome_arquivo = tela.lineEdit.text()

#final do programa
tela.show()
app.exec()
```
